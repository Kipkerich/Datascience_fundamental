## MACHINE LEARNING
The use of data as a learning point to make decision and predictions in order to solve real life problems

### Types of machine learning algorithm
Supervised- Uses labelled data to make decisions e.g linear and logistic regression
Unspervised- Algorithms where data is unlabelled eg Kmeans and K nearest neighbors
            - Clustering  is the most common method that uses this algorithm

##### Application of unspervised algorithm
1. Clustering
2. Anomaly networks
3. Neural networks

# Scaling
A strategy used to reduce noise and improve in the distribution structure of the data

## Algorithm that rely on scaling
1. Linear Regression
2. Logistic regression
3. Linear Discriminant Analaysis
4. Gaussian Naive Bayes
5. Neural networks
